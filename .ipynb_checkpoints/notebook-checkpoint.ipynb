```python
order_reviews_df.info()
```

- menangani missing value

```python
# baris data yang mengandung missing value pada column review_comment_title 
order_reviews_df[order_reviews_df.review_comment_title.isna()]
```

```python
# baris data yang mengandung missing value pada column review_comment_message 
order_reviews_df[order_reviews_df.review_comment_message.isna()]
```

dilihat bahwa baris data tersebut masih mengandung banyak informasi penting sehingga sayang jika langsung dibuang. Oleh karena itu, pada kasus ini, menggunakan metode imputation untuk menangani missing value.

karena jika menghapus semua missing value akan menghilangkan banyak informasi penting maka akan dilakukan perubahan nilai dengan nilai default string pada column review_comment_title berupa "No Title" dan pada column review_comment_message berupa "No Message"

```python
# mengubah missing value dengan nilai default
order_reviews_df["review_comment_title"].fillna("No Title", inplace=True)
order_reviews_df["review_comment_message"].fillna("No Message", inplace=True)
```

```python
#mengecek kembali jumlah missing value
order_reviews_df.isna().sum()
```

### Eksplorasi Data orders_df dan customers_df

```python
#merge data orders_df dan customers_df
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

### Eksplorasi products_df dan product_category_df

```python
# merge data products_df dan product_category_df
all_product_category = pd.merge(
    left= products_df,
    right = product_category_df,
    how = "left",
    left_on='product_category_name',
    right_on='product_category_name'
)
all_product_category.head()
```

```python
#merge order_item_df dan order_payments_df
order_payments_item = pd.merge(
    left = order_item_df,
    right = order_payments_df,
    how = "left",
    left_on='order_id',
    right_on='order_id'
)
order_payments_item.head()
```

#### tahap merge semua table

```python
merge1=pd.merge(
    left = orders_customers_df,
    right = order_payments_item,
    how = 'left',
    left_on='order_id',
    right_on='order_id'
)
merge1.head()
    
```

```python
merge2=pd.merge(
    left = merge1,
    right = all_product_category,
    how = 'left',
    left_on='product_id',
    right_on='product_id'
)
merge2.head()
```

```python
merge3=pd.merge(
    left = merge2,
    right = order_reviews_df,
    how = 'left',
    left_on='order_id',
    right_on='order_id'
)
merge3.head()
```

```python
all_df=pd.merge(
    left = merge3,
    right = sellers_df,
    how = 'left',
    left_on='seller_id',
    right_on='seller_id'
)
all_df.head()
```

### RFM Analysis

```python
all_df['order_purchase_timestamp'].max()
```

```python
all_df['order_purchase_timestamp'].min()
```

```python
import datetime as dt
current_time =dt.datetime(2018,12,10)
```

```python
all_df['order_purchase_timestamp'] = pd.to_datetime(all_df['order_purchase_timestamp'])

# Group by 'customer_id' and calculate Recency, Frequency, and Monetary
recency = (current_time - all_df.groupby('customer_id')['order_purchase_timestamp'].max()).dt.days
frequency = all_df.groupby('customer_id')['order_id'].count()
monetary = all_df.groupby('customer_id')['price'].sum()

# Create a new DataFrame with the calculated metrics
rfm = pd.DataFrame({
    'customer_id': recency.index,
    'Recency': recency.values,
    'Frequency': frequency.values,
    'Monetary': monetary.values
})

col_list = ['customer_id','Recency','Frequency','Monetary']
rfm.columns = col_list
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
 
colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]
 
# beri comentar pada ax[index].set_xticks([]) bila ingin melihat customer nya by id


sns.barplot(y="Recency", x="customer_id", data=rfm.sort_values(by="Recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel("customer_id")
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=90)
ax[0].set_xticks([])

 
sns.barplot(y="Frequency", x="customer_id", data=rfm.sort_values(by="Frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel('customer_id')
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].set_xticklabels(ax[0].get_xticklabels(), rotation=90)
ax[1].set_xticks([])

 
sns.barplot(y="Monetary", x="customer_id", data=rfm.sort_values(by="Monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel('customer_id')
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
ax[2].set_xticklabels(ax[0].get_xticklabels(), rotation=90)
ax[2].set_xticks([])

 
plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.show()
```
